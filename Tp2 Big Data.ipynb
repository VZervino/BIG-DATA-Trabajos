```python
import pandas as pd
```

```python
#Lo primero que hacemos es importar el dataframe .csv
df = pd.read_csv("Base Airbnb NY.csv")
df
```

```python
#a) Tengan en cuenta si hay valores duplicados y eliminenlos.
df = df.drop_duplicates()
df
```

```python
#b) Eliminen las columnas que no tienen informacion de interes.
columnas_eliminadas = ['name', 'host_name', 'last_review'] 
df = df.drop(columns=columnas_eliminadas)
df
```

```python
#c)
#Identificamos las filas que contienen missing values
missing_values = df[df.isna().any(axis=1)]
# Eliminamos las filas con los valores faltantes
df = df.dropna()
#Chequeamos que ninguna fila tenga almenos un valor restante
print(df.isnull().sum())

#FALTA JUSTIFICAR PORQUE HICIMOS ESTO
```

```python
#d) Tomamos una decision para las observaciones con outliers o valores que no tienen sentido.

# Utilizamos solo las columnas con valores numericos, no con valores string.
import matplotlib.pyplot as plt
import seaborn as sns

columnas_numericas = df.select_dtypes(include=['float64', 'int64']).columns

# Para ver bien los datos hacemos unn boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[columnas_numericas])
plt.title("Boxplot de Datos Numéricos")
plt.show()
#A partir del grafico vemos que als variables id y host_id tienen una alta variabilidad en los datos. Sin embargo, por la naturaleza de las variables, estas no son utiles para el analisis estadistico, por ende no nos precocupan. 

```

```python
#e) Transformamos las variables ‘neighbourhood group’ y ‘room type’ a variables numericas.
def asignar_valores_numericos(columna):
    mapeo = {}
    valor = 0
    for categoria in columna.unique():
        mapeo[categoria] = valor
        valor += 1
    return mapeo

# Primero obtenemos el  mapeo de valores numéricos para las dos variables: 'neighbourhood_group' y 'room_type'
mapeo_neighbourhood_group = asignar_valores_numericos(df['neighbourhood_group'])
mapeo_room_type = asignar_valores_numericos(df['room_type'])

# Despues ya reemplazamos los valores en la base de datos. 
df['neighbourhood_group'] = df['neighbourhood_group'].replace(mapeo_neighbourhood_group)
df['room_type'] = df['room_type'].replace(mapeo_room_type)

# Veamos las primeras 4 filas de la base con estas variables numericas para ver el cambio.
print(df.head())
#Observamos que se transformaron las dos columnas de variables exigidas a variables numericas: 'neighbourhood_group' y 'room_type'.
```

```python
#f) Usamos los comandos groupby y join o merge, para crear una columna que tenga la cantidad de oferentes por “Neighbourhood group”.
#La llamamos offer group.
# Contamos  la cantidad de oferentes por 'Neighbourhood group' con groupby
oferentes_por_grupo = df.groupby('neighbourhood_group')['host_id'].count().reset_index()
oferentes_por_grupo.columns = ['neighbourhood_group', 'offer_group']

# Cambiamos el nombre de la nueva columna a 'offer_group' como lo exige la consigna
oferentes_por_grupo.rename(columns={'offer_group': 'offer_group'}, inplace=True)

# Combinamos la nueva columna con la base de datos original usando merge
df = pd.merge(df, oferentes_por_grupo, on='neighbourhood_group', how='left')

# Lo mostramos 
print(df.head())
```
