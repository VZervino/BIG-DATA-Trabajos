Parte I: 

Limpieza de la base de datos

```python
import pandas as pd
```

```python
#Lo primero que hacemos es importar el dataframe .csv
df = pd.read_csv("Base Airbnb NY.csv")
df
```

```python
#a) Tengan en cuenta si hay valores duplicados y eliminenlos.
df = df.drop_duplicates()
df
```

```python
#b) Eliminen las columnas que no tienen informacion de interes.
columnas_eliminadas = ['name', 'host_name', 'last_review'] 
df = df.drop(columns=columnas_eliminadas)
df
```

```python
#c)
#Identificamos las filas que contienen missing values
missing_values = df[df.isna().any(axis=1)]
# Eliminamos las filas con los valores faltantes
df = df.dropna()
#Chequeamos que ninguna fila tenga almenos un valor restante
print(df.isnull().sum())
```

```python
#d) Tomamos una decision para las observaciones con outliers o valores que no tienen sentido.

# Utilizamos solo las columnas con valores numericos, no con valores string.
import matplotlib.pyplot as plt
import seaborn as sns

columnas_numericas = df.select_dtypes(include=['float64', 'int64']).columns

# Para ver bien los datos hacemos unn boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[columnas_numericas])
plt.title("Boxplot de Datos Numéricos")
plt.show()
#A partir del grafico vemos que als variables id y host_id tienen una alta variabilidad en los datos. Sin embargo, por la naturaleza de las variables, estas no son utiles para el analisis estadistico, por ende no nos precocupan. 

```

```python
#e) Transformamos las variables ‘neighbourhood group’ y ‘room type’ a variables numericas.
def asignar_valores_numericos(columna):
    mapeo = {}
    valor = 0
    for categoria in columna.unique():
        mapeo[categoria] = valor
        valor += 1
    return mapeo

# Primero obtenemos el  mapeo de valores numéricos para las dos variables: 'neighbourhood_group' y 'room_type'
mapeo_neighbourhood_group = asignar_valores_numericos(df['neighbourhood_group'])
mapeo_room_type = asignar_valores_numericos(df['room_type'])

# Despues ya reemplazamos los valores en la base de datos. 
df['neighbourhood_group'] = df['neighbourhood_group'].replace(mapeo_neighbourhood_group)
df['room_type'] = df['room_type'].replace(mapeo_room_type)

# Veamos las primeras 4 filas de la base con estas variables numericas para ver el cambio.
print(df.head())
#Observamos que se transformaron las dos columnas de variables exigidas a variables numericas: 'neighbourhood_group' y 'room_type'.
```

```python
#f) Usamos los comandos groupby y join o merge, para crear una columna que tenga la cantidad de oferentes por “Neighbourhood group”.
#La llamamos offer group.
# Contamos  la cantidad de oferentes por 'Neighbourhood group' con groupby
oferentes_por_grupo = df.groupby('neighbourhood_group')['host_id'].count().reset_index()
oferentes_por_grupo.columns = ['neighbourhood_group', 'offer_group']

# Cambiamos el nombre de la nueva columna a 'offer_group' como lo exige la consigna
oferentes_por_grupo.rename(columns={'offer_group': 'offer_group'}, inplace=True)

# Combinamos la nueva columna con la base de datos original usando merge
df = pd.merge(df, oferentes_por_grupo, on='neighbourhood_group', how='left')

# Lo mostramos 
print(df.head())
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#Elección de las variables
variables_PCA = ['latitude', 'longitude', 'price', 'minimum_nights', 
            'number_of_reviews', 'reviews_per_month', 
            'calculated_host_listings_count', 'availability_365']

#Estandarización
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[variables_PCA])

#PCA
pca = PCA(n_components=2)
pca.fit(df_scaled)
df_pca = pca.transform(df_scaled)

#Componentes guardados en un dataframe
pca_df = pd.DataFrame(data=df_pca, columns=['PC1', 'PC2'])
pca_df[['neighbourhood_group', 'room_type']] = df[['neighbourhood_group', 'room_type']]

print(pca_df)

#Gráfico
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2')
plt.xlabel('Componente Principal 1 (PC1)')
plt.ylabel('Componente Principal 2 (PC2)')
plt.title('PCA: Variables en dos dimensiones')
plt.show()

# Porcentaje de varianza explicada por los dos componentes
explained_variance_ratio = pca.explained_variance_ratio_
print("Porcentaje de varianza explicada por PC1 y PC2:")
print("PC1:", explained_variance_ratio[0])
print("PC2:", explained_variance_ratio[1])

# Loadings de los componentes principales
loadings = pca.components_
loadings_df = pd.DataFrame(loadings, columns=variables_PCA, index=['PC1', 'PC2'])
print("\nLoadings de los componentes principales:")
print(loadings_df)
```

```python
#La varianza explicada por los primeros 2 componentes es del 37,7%. Los loadings muestran que el componente PC1 explica variabilidad relacionada con el número de reviews y el número por mes, junto con la disponibilidad 365 días. Esto tiene sentido porque las primeras 2 están lógicamente muy relacionadas entre si y el promedio por mes tiene relación obviamente con si la disponibilidad es todos los días del año. Variables como la latitud, longitud, precio, cantidad mínima de noches y conteo de hosts no están tan relacionadas con el componente 1. El componente 2 también agarra variabilidad relacionada con la disponibilidad 365 días pero que se relaciona aparte con el precio, la cantidad mínima de noches y la cantidad de hosts. El método PCA con 2 variables deja prácticamente sin explicar la variabilidad de la latitud, ya que ambos coeficientes son bajos en valor absoluto. El gráfico no muestra una relación clara entre los dos coeficientes principales, lo que tiene sentido porque los componentes son ortogonales.
```
